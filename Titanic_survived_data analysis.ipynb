# Some insights show through charts and graphs

```python
#show passenger count by class
import matplotlib.pyplot as plt

classes = ['1st', '2nd', '3rd']
passenger_count = [216, 184, 491]

plt.bar(classes, passenger_count,edgecolor="#2F3C7E",color="#3783a3",width=0.5,)
plt.xlabel('Passenger Class')
plt.ylabel('Passenger Count')
plt.title('Passenger Count by Class')
plt.show()
```

```python
# Calculate passenger count by gender
gender_count = df['Sex'].value_counts()
gender_count
```

```python
#show passenger count by gender

colors = ['#5e7e8c', 'pink']

plt.figure(figsize=(6, 6))
plt.pie(gender_count, labels=gender_count.index, startangle=140,autopct='%1.1f%%',colors=colors)
plt.title('Distribution of Passengers by Gender')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

import matplotlib.pyplot as plt


```

```python
age_data = df['Age']

# Create a histogram
plt.figure(figsize=(8, 6))
plt.hist(age_data, bins=20, edgecolor='k', color='skyblue')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Passenger Ages')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

# Create a scatter plot
plt.figure(figsize=(17, 10))
plt.scatter(df['Age'],df['Fare'], color='green', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs. Fare')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python


# Create a box plot
plt.figure(figsize=(7, 6))
plt.boxplot([df[df['Pclass'] == 1]['Fare'], df[df['Pclass'] == 2]['Fare'], df[df['Pclass'] == 3]['Fare']], labels=['1st Class', '2nd Class', '3rd Class'])
plt.ylabel('Fare')
plt.title('Box Plot of Fare by Passenger Class')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#In Pandas df.unstack() function reshapes the given DataFrame by converting the row index to a column label
class_survival = df.groupby(['Pclass', 'Survived']).size().unstack()
print(class_survival)
# Create a stacked bar chart
class_survival.plot(kind='bar', stacked=True, color=['#ADD8E6', '#00008B'], alpha=0.7)

plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Survival Status by Passenger Class')
plt.legend(['Did Not Survive', 'Survived'], loc='upper left')
plt.show()
```

```python

# Count the number of passengers from each embarkation point
embarkation_count = df['Embarked'].value_counts()

# Create a horizontal bar chart
plt.figure(figsize=(8, 6))
embarkation_count.plot(kind='barh', color='#628ba6')
plt.xlabel('Count')
plt.ylabel('Embarkation Point')
plt.title('Passenger Distribution by Embarkation Point')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
